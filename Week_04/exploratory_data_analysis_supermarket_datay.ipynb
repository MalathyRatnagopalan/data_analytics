## Libraries and settings

```python
# Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Read and select variables
df = pd.read_csv("supermarkets_data_enriched.csv") [[ 'id',
                                                     'brand',
                                                     'lat',
                                                     'lon']]

# Subset
df_sub = df.loc[df['brand'].isin(['Coop', 'Migros'])]
df_sub

# Colors
colors = {'Coop': 'green', 'Migros': 'orange'}

# Plot
plt.scatter(df_sub['lon'],
            df_sub['lat'],
            c=df_sub['brand'].map(colors))

# Count the number of supermarkets per brand
brand_counts = df['brand'].value_counts()
print(brand_counts)

# Bar chart
plt.figure(figsize=(8, 6))
brand_counts.plot(kind='bar', color=['orange', 'green'], alpha=0.7)
plt.title('Number of Supermarkets per Brand')
plt.xlabel('Brand')
plt.ylabel('Count')
plt.show()

# Select numeric variables for the scatterplot matrix
numeric_vars = ['lat', 'lon']

# Create scatterplot matrix
scatterplot_matrix = sns.PairGrid(df[numeric_vars])
scatterplot_matrix.map_upper(sns.scatterplot)
scatterplot_matrix.map_diag(sns.histplot)
scatterplot_matrix.add_legend()
plt.show()

# Subset for Denner, Volg, and Landi
df_subset_locations = df.loc[df['brand'].isin(['Denner', 'Volg', 'Landi'])]

# Colors for different brands
colors_subset = {'Denner': 'red', 'Volg': 'blue', 'Landi': 'purple'}

# Plot locations with different colors for brands
plt.scatter(df_subset_locations['lon'], df_subset_locations['lat'],
            c=df_subset_locations['brand'].map(colors_subset),
            alpha=0.7)
plt.title('Supermarket Locations by Brand')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.legend(title='Brand')
plt.show()
```
